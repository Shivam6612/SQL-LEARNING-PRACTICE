Dump all the data into car table

```python
import csv
with open('car.data','r') as f:
    car_data=csv.reader(f)
    for i in car_data:
        print(i)
        #inserting the data
        values = ",".join([f"'{x}'" for x in i])
        cur.execute(f'insert into cardataset.car values({values})')
```
